```python
class BustaPaga:
    def __init__(self):
        # Dati principali
        self.retribuzione_tabellare = 8.13
        self.indennita_settore = 0.21
        self.straordinario_forfait = 1.45
        self.acconto_futuri_aumenti = 0.53
        self.ferie_residui = 13.33
        self.permessi_riduzione_orario = 3.33
        self.permessi_ex_festivita = 8.67
        self.ore_lavorate = 160.0
        self.festivita = 16
        self.gratifica_natalizia = 148.88
        self.rata_addizionale_regionale = 0.34
        
        # Competenze
        self.totale_competenze = 1966.45
        self.irpef_lorda = 439.98
        self.detrazioni_lavoratore = 30.00
        self.irpef_netta = 409.98
        self.totale_trattenute = 463.81
        self.contributo_inps = 180.68
        self.contributo_inps_esonero_ivs = 127.19
        self.totale_contributi = 53.49
        self.arr_precedente = 0.23
        self.arr_attuale = 0.59
        self.totale_spettante = 1966.11

    def set_rettifica(self, **kwargs):
        chiavi_valide = {
            'retribuzione_tabellare': float,
            'indennita_settore': float,
            'straordinario_forfait': float,
            'acconto_futuri_aumenti': float,
            'ferie_residui': float,
            'permessi_riduzione_orario': float,
            'permessi_ex_festivita': float,
            'ore_lavorate': float,
            'festivita': float,
            'gratifica_natalizia': float,
            'rata_addizionale_regionale': float,
            'totale_competenze': float,
            'irpef_lorda': float,
            'detrazioni_lavoratore': float,
            'irpef_netta': float,
            'totale_trattenute': float,
            'contributo_inps': float,
            'contributo_inps_esonero_ivs': float,
            'totale_contributi': float,
            'arr_precedente': float,
            'arr_attuale': float,
            'totale_spettante': float
        }

        for chiave, valore in kwargs.items():
            if chiave in chiavi_valide:
                try:
                    setattr(self, chiave, chiavi_valide[chiave](valore))
                except ValueError:
                    print(f"Errore: il valore per '{chiave}' deve essere di tipo {chiavi_valide[chiave].__name__}.")
            else:
                print(f"Errore: '{chiave}' non è un attributo valido.")

    def calcola_straordinari(self):
        try:
            ore_straordinarie = float(input("Inserisci il numero di ore di straordinario effettuate: "))
        except ValueError:
            print("Errore: per favore inserisci un numero valido.")
            return 0.0

        tariffa_straordinario = self.retribuzione_tabellare * self.straordinario_forfait
        compenso_straordinario = ore_straordinarie * tariffa_straordinario
        self.totale_competenze += round(compenso_straordinario, 2)
        return round(compenso_straordinario, 2)

    def calcola_netto_verticale(self):
        netto_in_busta = self.totale_spettante - (self.irpef_netta + self.totale_contributi) + (self.arr_precedente + self.arr_attuale)
        return round(netto_in_busta, 2)

    def calcola_netto_orizzontale(self):
        netto_confronto = self.totale_competenze - (self.totale_trattenute + self.arr_precedente + self.arr_attuale)
        return round(netto_confronto, 2)
    
    def calcola_netto_semplice(self):
        netto_semplice = self.totale_competenze - self.totale_trattenute
        return round(netto_semplice, 2)
    
    def calcola_netto_differenziale(self):
        differenza = self.totale_competenze - (self.irpef_netta + self.totale_contributi)
        netto_differenziale = differenza + (self.arr_precedente + self.arr_attuale)
        return round(netto_differenziale, 2)
    
    def calcola_netto_dettaglio(self):
        netto_dettaglio = (self.totale_spettante 
                           - self.irpef_netta 
                           - self.totale_contributi 
                           + self.arr_precedente 
                           + self.arr_attuale)
        return round(netto_dettaglio, 2)

    def stampa_busta_paga(self):
        netto = self.calcola_netto_verticale()
        netto_confronto = self.calcola_netto_orizzontale()
        netto_semplice = self.calcola_netto_semplice()
        netto_differenziale = self.calcola_netto_differenziale()
        netto_dettaglio = self.calcola_netto_dettaglio()
        differenza_netto = netto_semplice - netto_dettaglio
        
        print(f"{'='*40}")
        print(f"{'Busta Paga':^40}")
        print(f"{'='*40}")
        print(f"Retribuzione Tabellare: {self.retribuzione_tabellare:.2f} €")
        print(f"Indennità Settore: {self.indennita_settore:.2f} €")
        print(f"Straordinario Forfait: {self.straordinario_forfait:.2f} €")
        print(f"Acconto Futuri Aumenti: {self.acconto_futuri_aumenti:.2f} €")
        print(f"Ferie Residue: {self.ferie_residui:.2f} ore")
        print(f"Permessi Riduzione Orario Residui: {self.permessi_riduzione_orario:.2f} ore")
        print(f"Permessi Ex Festività Residui: {self.permessi_ex_festivita:.2f} ore")
        print(f"Ore Lavorate: {self.ore_lavorate:.2f} ore")
        print(f"Gratifica Natalizia: {self.gratifica_natalizia:.2f} €")
        print(f"Rata Addizionale Regionale: {self.rata_addizionale_regionale:.2f} €")
        print(f"{'-'*40}")
        print(f"Totale Competenze: {self.totale_competenze:.2f} €")
        print(f"IRPEF Lorda: {self.irpef_lorda:.2f} €")
        print(f"Detrazioni Lavoratore: {self.detrazioni_lavoratore:.2f} €")
        print(f"IRPEF Netta: {self.irpef_netta:.2f} €")
        print(f"Contributo INPS: {self.contributo_inps:.2f} €")
        print(f"Contributo INPS Esonero IVS: {self.contributo_inps_esonero_ivs:.2f} €")
        print(f"Totale Contributi: {self.totale_contributi:.2f} €")
        print(f"Totale Spettante: {self.totale_spettante:.2f} €")
        print(f"Arretrati Precedenti: {self.arr_precedente:.2f} €")
        print(f"Arretrati Attuali: {self.arr_attuale:.2f} €")
        print(f"{'='*40}")
        print(f"Netto in Busta (Verticale): {netto:.2f} €")
        print(f"Netto Dettaglio (Orizontale): {netto_dettaglio:.2f} €")
        print(f"{'='*40}")
        print(f"{'Analisi Di Confronto Discrepanze':^40}")
        print(f"{'='*40}")
        print(f"Netto Confronto (Orizzontale): {netto_confronto:.2f} €")
        print(f"Netto Semplice: {netto_semplice:.2f} €")
        print(f"Netto Differenziale: {netto_differenziale:.2f} €")
        print(f"{'-'*40}")
        print(f"{'Mancanza':^40} \n Differenza Netto: {differenza_netto:.2f} €")


# Esempio di utilizzo
busta = BustaPaga()

# Esegui i calcoli dopo aver chiesto il valore input
busta.calcola_straordinari()

# Stampa la busta 
busta.stampa_busta_paga()
```
